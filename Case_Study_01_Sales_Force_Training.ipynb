# DataFrame

```python
import pandas as pd
from IPython.display import display

# Data transaksi
data = [
    100, 150, 50, 100, 130, 120, 100, 85, 70, 150,
    150, 120, 50, 100, 100, 140, 90, 150, 50, 90,
    120, 100, 110, 75, 65
]

# Membuat DataFrame
df = pd.DataFrame({'TransactionAmount': data})

# print(df)
display(df)




```

# Measure Of Central Tendency

```python
# Mean
mean_val = df['TransactionAmount'].mean()

# Median
median_val = df['TransactionAmount'].median()

# Mode
mode_val = df['TransactionAmount'].mode().tolist()

print(f"Mean   : {mean_val}")
print(f"Median : {median_val}")
print(f"Mode   : {mode_val}")
```

# Measure Of Variability

```python
# Max
max_val = df['TransactionAmount'].max()

# Min
min_val = df['TransactionAmount'].min()


# Range
range_val = df['TransactionAmount'].max() - df['TransactionAmount'].min()

# Variance
variance_val = df['TransactionAmount'].var()

# Standard Deviation
std_dev_val = df['TransactionAmount'].std()

# IQR
q1 = df['TransactionAmount'].quantile(0.25)
q3 = df['TransactionAmount'].quantile(0.75)
iqr_val = q3 - q1

print(f"Max                : {max_val}")
print(f"Min                : {min_val}")
print(f"Range              : {range_val}")
print(f"Variance           : {variance_val}")
print(f"Standard Deviation : {std_dev_val}")
print(f"IQR                : {iqr_val}")
```

# H0 & H1

```python

from scipy import stats

# One-sample t-test terhadap nilai acuan 100
t_stat, p_val = stats.ttest_1samp(df['TransactionAmount'], 100)

# Keputusan
alpha = 0.05
keputusan = "Tolak H0" if p_val < alpha else "Gagal Tolak H0"

# Hasil
hasil_ttest = pd.DataFrame({
    't-statistic': [t_stat],
    'p-value': [p_val],
    'alpha': [alpha],
    'Keputusan': [keputusan]
})


# print()
# caas_jupyter_tools.display_dataframe_to_user(name="Hasil One-Sample T-Test TransactionAmount", dataframe=hasil_ttest)
from IPython.display import display
display(hasil_ttest)
```

# T-Test pada data rata-rata jumlah transaksi (dengan alpha = 5%)

```python
import pandas as pd
from scipy import stats

# Data transaksi
data = [
    100, 150, 50, 100, 130, 120, 100, 85, 70, 150,
    150, 120, 50, 100, 100, 140, 90, 150, 50, 90,
    120, 100, 110, 75, 65
]
df = pd.DataFrame({'TransactionAmount': data})

# Nilai acuan (mu0)
mu0 = 100

# One-Sample T-Test
t_stat, p_val = stats.ttest_1samp(df['TransactionAmount'], mu0)

# Alpha
alpha = 0.05

# Keputusan
if p_val < alpha:
    keputusan = "Tolak H0 (rata-rata berbeda signifikan dari 100)"
else:
    keputusan = "Gagal Tolak H0 (tidak ada perbedaan signifikan dari 100)"

# Hasil dalam DataFrame
hasil_ttest = pd.DataFrame({
    't-statistic': [t_stat],
    'p-value': [p_val],
    'alpha': [alpha],
    'Keputusan': [keputusan]
})

hasil_ttest
```
